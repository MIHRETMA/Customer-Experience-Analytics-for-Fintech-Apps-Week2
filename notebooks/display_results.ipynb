```python
def run_query(sql): # Execute a SQL query and return the results as a DataFrame
    return pd.read_sql(sql, conn)

sql = "SELECT * FROM bank_schema.banks;"

run_query(sql)
```

```python
sql = "SELECT * FROM bank_schema.reviews;"

run_query(sql)
```

```python
# Average rating per bank

sql = """
SELECT r.bank_name, ROUND(AVG(v.rating), 2) AS avg_rating
FROM bank_schema.reviews v
JOIN bank_schema.banks r ON r.bank_id = v.bank_id
GROUP BY r.bank_name;
"""

run_query(sql)
```

```python
# Count reviews per banks ensuring all 400 reviews are inserted for each bank

sql = """
SELECT r.bank_name, COUNT(*) AS total_reviews
FROM bank_schema.reviews v
JOIN bank_schema.banks r ON r.bank_id = v.bank_id
GROUP BY r.bank_name;
"""

run_query(sql)
```
